```python
#Projeto: Análise Exploratória e Estratégica - Brasileirão 2009.

#Este projeto teve como objetivo transformar dados brutos das 380 partidas do Campeonato Brasileiro de 2009 em insights de desempenho e engajamento.
#A análise percorreu desde a limpeza de dados até a criação de métricas de eficiência de torcida e arbitragem.
```

```python
#Resumo Técnico 

#Python & Pandas: Limpeza de dados (Data Cleaning) e criação de tabelas de classificação do zero.
#Feature Engineering: Desenvolvimento de colunas como Evolução por Turno, Dependência de Casa e Público por Ponto.
#Data Visualization: Gráficos avançados com Seaborn e Matplotlib para identificar padrões de comportamento de público e arbitragem.
```

```python
#Principais Insights 

#A Arrancada do Título
#O campeão (Flamengo) demonstrou que o título foi conquistado na recuperação. Enquanto alguns times estagnaram, a evolução do 1º para o 2º turno foi o diferencial estatístico.

#Engajamento e Fidelidade
#O gráfico de Público por Ponto revelou que as maiores torcidas (Flamengo e Atlético-MG) mantiveram médias altíssimas, provando que o engajamento em 2009 foi um motor financeiro e motivacional, independentemente da posição momentânea na tabela.

#Oportunidades de Negócio (Horários)
#Identificamos que o horário de Domingo às 16h possui uma média de público superior aos jogos de meio de semana, validando a estratégia de "horário nobre" para grandes rendas.

#Rigor na Arbitragem
#O árbitro Sandro Meira Ricci liderou o ranking de cartões vermelhos com [15] expulsões, destacando-se como o perfil mais rigoroso da temporada.
```
