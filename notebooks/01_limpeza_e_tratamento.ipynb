```python
#Etapa 1: Limpeza e Padronizacao.

import pandas as pd
import matplotlib as plt
import numpy as np
import os
```

```python
df = pd.read_csv("../data/brasileirao_2009_add.csv")
df
#Abrindo pela primeira vez o csv em python visualizei que esqueci de remover algumas colunas em branco que no momento nao serao tao relevantes para a analise.
```

```python
#Criar uma lista para remover as colunas.
colunas_remocao = [
        "formacao_mandante",
        "formacao_visitante",
        "tecnico_mandante",
        "tecnico_visitante"
]
df.drop(columns=colunas_remocao, inplace=True)
#Atuais Colunas apos a remocao.
print("Colunas Atuais:", df.columns)
df.head()
#Durante a inspeção inicial, identifiquei colunas com 100% de valores ausentes (NaN). 
#Para otimizar o consumo de memória e focar nas variáveis relevantes para a análise de público e desempenho, optei por remover estas colunas do dataset de trabalho.
```

```python
#Novamente identifiquei algo que pode me complicar futuramente, o "-" que seria o Empate na coluna vencedor
df["vencedor"] = df['vencedor'].str.strip().replace("-", "Empate")
df.head()
```

```python
#Apenas para fazer um teste na coluna publico_jogo eu tentei somar todas os publicos e deu um erro, que eu ja previa por conta do python entao trataremos esses dados para que sejam de formato numerico para que a manipulacao seja feita
#ID 3082 e ID 3086
df["publico_jogo"] = df["publico_jogo"].astype(str)
df["publico_jogo"] = df["publico_jogo"].str.replace("....", "0", regex=False)
df["publico_jogo"] = df["publico_jogo"].str.replace(".", "", regex=False)
df["publico_jogo"] = pd.to_numeric(df["publico_jogo"], errors="coerce").fillna(0).astype(int) #corrijindo o erro

#teste

total_publico = df["publico_jogo"].sum()
media_publico = df["publico_jogo"].mean()

print(f"Soma Total correta: {df['publico_jogo'].sum():,}".replace(",", "."))
print(f"Média correta: {df['publico_jogo'].mean():.2f}")

#Final da limpeza e tratamento dos dados
```
